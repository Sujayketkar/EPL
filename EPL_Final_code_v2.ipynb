```python
import pandas as pd
import re

# Function to convert short year format to full year format
def convert_season(season):
    match = re.match(r'(\d{4})-(\d{2})', season)
    if match:
        year_start = match.group(1)
        year_end = match.group(2)
        year_end_full = '19' + year_end if int(year_end) < 100 else '20' + year_end
        return f'{year_start}-{year_end_full}'
    return season



# Apply the conversion function to the 'Season' column
df_2['Season'] = df_2['Season'].apply(convert_season)


```

```python
import dash
from dash import dcc, html
import dash_table
import pandas as pd
from dash.dependencies import Input, Output

# Team colors dictionary (for reference)
team_colors = {
    'Arsenal': '#EF0107',
    'Aston Villa': '#95BFE5',
    'Chelsea': '#034694',
    'Liverpool': '#C8102E',
    'Man City': '#6CABDD',
    'Newcastle': '#241F20',
    'Oldham': '#004D98',
    'Sheffield United': '#EE2737',
    'Southampton': '#D71920',
    'West Ham': '#7A263A',
    'Norwich': '#FFF200',
    'Tottenham': '#132257',
    'Everton': '#003399',
    'Ipswich': '#003B8F',
    'Leeds': '#FFCD00',
    'Wimbledon': '#003366',
    'Blackburn': '#1B458F',
    'Coventry': '#00A9CE',
    'Man United': '#DA291C',
    'QPR': '#1C2C5B',
    'Sheffield Weds': '#0073D0',
    'Swindon': '#FF0000',
    'Crystal Palace': '#1B458F',
    'Leicester': '#0053A0',
    "Nott'm Forest": '#DD0000',
    'Bolton': '#FFFFFF',
    'Middlesbrough': '#DA291C',
    'Derby': '#FFFFFF',
    'Sunderland': '#EB172B',
    'Barnsley': '#E30613',
    'Charlton': '#C8102E',
    'Watford': '#FBEE23',
    'Bradford': '#7A003C',
    'Fulham': '#FFFFFF',
    'Birmingham': '#0033A1',
    'West Brom': '#122F67',
    'Portsmouth': '#003087',
    'Wolves': '#FDB913',
    'Wigan': '#005BAC',
    'Reading': '#003087',
    'Hull': '#F4A300',
    'Stoke': '#E03A3E',
    'Burnley': '#6C1D45',
    'Blackpool': '#FF8300',
    'Swansea': '#FFFFFF',
    'Cardiff': '#0070B5',
    'Bournemouth': '#DA291C',
    'Brighton': '#0057B8',
    'Huddersfield': '#0073CF',
    'Brentford': '#D00027',
    'Luton': '#FF4A00'
}

# Initialize the app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div(style={'backgroundColor': '#3d195b', 'padding': '20px', 'fontSize': '18px'}, children=[
    html.H1("EPL Stats Central: From 1993 to 2024", style={'textAlign': 'center', 'color': '#FFFFFF', 'fontSize': '24px', 'fontWeight': 'bold'}),
    
    html.Div([
        html.Label('Select Season:', style={'color': '#FFFFFF'}),
        dcc.Dropdown(
            id='season-dropdown',
            options=[{'label': season, 'value': season} for season in df_1['Season'].unique()],
            value=df_1['Season'].unique()[0],  # Default value
            clearable=False
        )
    ], style={'width': '20%', 'display': 'inline-block', 'verticalAlign': 'top'}),
    
    html.Div(id='winner-container', style={'paddingTop': '20px', 'width': '20%', 'display': 'inline-block', 'verticalAlign': 'top'}),
    
    html.Div(id='titles-container', style={'paddingTop': '20px', 'width': '20%', 'display': 'inline-block', 'verticalAlign': 'top'}),
    
    html.Div(id='table-container', style={'paddingTop': '20px', 'width': '60%', 'margin': '0 auto'})
])

# Callback to update table, league winner, and number of titles based on season selection
@app.callback(
    [Output('table-container', 'children'),
     Output('winner-container', 'children'),
     Output('titles-container', 'children')],
    [Input('season-dropdown', 'value')]
)
def update_table_and_winner(selected_season):
    # Filter the data for the selected season
    df_season = df_1[df_1['Season'] == selected_season]
    
    # Create a dataframe for team statistics
    teams = df_season['HomeTeam'].unique()
    
    team_stats = []
    for team in teams:
        home_games = df_season[df_season['HomeTeam'] == team]
        away_games = df_season[df_season['AwayTeam'] == team]
        total_games = pd.concat([home_games, away_games])

        wins = len(total_games[(total_games['FullTimeResult'] == 'H') & (total_games['HomeTeam'] == team)]) + \
               len(total_games[(total_games['FullTimeResult'] == 'A') & (total_games['AwayTeam'] == team)])

        losses = len(total_games[(total_games['FullTimeResult'] == 'A') & (total_games['HomeTeam'] == team)]) + \
                 len(total_games[(total_games['FullTimeResult'] == 'H') & (total_games['AwayTeam'] == team)])

        draws = len(total_games[total_games['FullTimeResult'] == 'D'])

        goals_scored = home_games['FullTimeHomeTeamGoals'].sum() + away_games['FullTimeAwayTeamGoals'].sum()
        goals_conceded = home_games['FullTimeAwayTeamGoals'].sum() + away_games['FullTimeHomeTeamGoals'].sum()

        # Calculate total points (home and away)
        points = home_games['HomeTeamPoints'].sum() + away_games['AwayTeamPoints'].sum()

        team_stats.append({
            'Team': team,
            'Games_Played': len(total_games),
            'Wins': wins,
            'Losses': losses,
            'Draws': draws,
            'Goals_Scored': goals_scored,
            'Goals_Conceded': goals_conceded,
            'GD': goals_scored - goals_conceded,
            'Points': points
        })

    df_team_stats = pd.DataFrame(team_stats).sort_values(by=['Points'], ascending=False)
    
    # Create a table with italicized text except for the team names
    table = dash_table.DataTable(
        columns=[
            {'name': 'Team', 'id': 'Team'},
            {'name': 'Games Played', 'id': 'Games_Played'},
            {'name': 'Wins', 'id': 'Wins'},
            {'name': 'Losses', 'id': 'Losses'},
            {'name': 'Draws', 'id': 'Draws'},
            {'name': 'Goals Scored', 'id': 'Goals_Scored'},
            {'name': 'Goals Conceded', 'id': 'Goals_Conceded'},
            {'name': 'GD', 'id': 'GD'},
            {'name': 'Points', 'id': 'Points'}
        ],
        data=df_team_stats.to_dict('records'),
        style_data_conditional=[
            # Italicize all columns except the Team column
            {
                'if': {'column_id': col},
                'fontStyle': 'italic'
            } for col in df_team_stats.columns if col != 'Team'
        ] + [
            # Apply team colors to the Team column
            {
                'if': {'filter_query': f'{{Team}} = "{team}"', 'column_id': 'Team'},
                'backgroundColor': team_colors.get(team, '#FFFFFF'),
                'color': 'white',
                'fontStyle': 'normal'
            } for team in team_colors
        ],
        style_header={'backgroundColor': 'black', 'color': 'white'},
        style_cell={'textAlign': 'center', 'backgroundColor': '#3d195b', 'color': '#FFFFFF'},
        style_table={'overflowX': 'auto', 'margin': '0 auto'}
    )
    
    # Fetch the league winner for the selected season
    winner_row = df_2[df_2['Season'] == selected_season]
    if not winner_row.empty:
        winner_team = winner_row['League_Winner'].values[0]
        # Calculate the number of titles won by the winner
        titles_won = df_2[df_2['League_Winner'] == winner_team].shape[0]
        # Use a default color for the winner box if team color is not available
        default_color = '#000000'
        winner_color = team_colors.get(winner_team, default_color)
        winner_box = html.Div(
            children=[
                html.H3(winner_team, style={'color': winner_color, 'fontSize': '24px', 'fontWeight': 'bold'})
            ],
            style={'backgroundColor': '#FFFFFF', 'color': winner_color, 'padding': '20px', 'borderRadius': '5px', 'textAlign': 'center'}
        )
        titles_box = html.Div(
            children=[
                html.H3(f"Titles Won: {titles_won}", style={'color': winner_color, 'fontSize': '24px', 'fontWeight': 'bold'})
            ],
            style={'backgroundColor': '#FFFFFF', 'color': winner_color, 'padding': '20px', 'borderRadius': '5px', 'textAlign': 'center'}
        )
    else:
        winner_box = html.Div(
            children=[html.H3('No Winner Available')],
            style={'backgroundColor': '#FFFFFF', 'color': '#000000', 'padding': '20px', 'borderRadius': '5px', 'textAlign': 'center'}
        )
        titles_box = html.Div(
            children=[html.H3('Titles Won: N/A')],
            style={'backgroundColor': '#FFFFFF', 'color': '#000000', 'padding': '20px', 'borderRadius': '5px', 'textAlign': 'center'}
        )

    return table, winner_box, titles_box

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True, port=3060)
```
